```python
%matplotlib inline
import matplotlib.pyplot as plt
import glob
from natsort import natsorted
```

```python
images = natsorted(glob.glob('./t/*'))

fig, axs = plt.subplots(nrows=5, ncols=4, figsize=(40,50), gridspec_kw={'wspace':.0, 'hspace':.0})
for ax, im in zip(axs.flatten(), images):
    ax.imshow(plt.imread(im))
    ax.set_xticks([])
    ax.set_yticks([])
    for sp in ax.spines.values():
        sp.set_visible(False)
fig.savefig('./static-images/fashion-9-scaling-img.png')
```
